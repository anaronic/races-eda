```python
# data cleanup
# only USA races for now, 50K or 50 miles, 2020
# step 1: show 50Mi or 50k
# represented as 50mi and 50km
```

```python
df[df['Event distance/length'] == '50mi']
```

```python
# combine 50k/50mi with isin 
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
#2020 only
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]
```

```python
#USA only - all filters together
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & ((df['Event name'].str.split('(')).str.get(1) == 'USA)')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & ((df['Event name'].str.split('(')).str.get(1) == 'USA)')]
df2.head()
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
df2['athlete_age'] = 2020-df2['Athlete year of birth']
```

```python
df2.head()
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis=1)
```

```python
df2.head()
```

```python
# cleanup null values
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
df2.reset_index(drop = True)
```

```python
# fixing types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
df2['event_name'] = df2['Event name'].astype(str)
df2['athlete_gender'] = df2['Athlete gender'].astype(str)
df2['avg_speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.head()
```

```python
# dropping old columns
df2 = df2.drop(['Athlete gender','Event name','Athlete average speed'],axis=1)
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
df2 = df2.rename(columns = {
    'Year of event':'year',
    'race_date':'race_day',
    'event_name':'race_name',
    'Event distance/length':'race_length',     
    'Event number of finishers':'race_number_of_finishers', 
    'Athlete performance':'athlete_performance',           
    'Athlete ID':'athlete_id',
    'avg_speed':'athlete_average_speed'
})
```

```python
df2.dtypes
```

```python
# reorder columns
df3 = df2[['race_day','race_name',"race_length",'race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```
